```python
# Standarize column names

data.columns = [col_name.lower().replace(' ', '_') for col_name in data.columns]
data.head()
```

## Exclude the following variables by looking at the definitions. Create a new empty list called drop_list. We will append this list and then drop all the columns in this list later:

- OSOURCE - symbol definitions not provided, too many categories
- ZIP CODE - we are including state already

```python
categorical.shape
```

```python
data["osource"].dtype
```

```python
data["zip"].dtype
```

```python
# Define the columns to be excluded
exclude_columns = ['osource', 'zip']

# Create an empty list to store columns to be dropped
drop_list = []

# Iterate through the exclude_columns list
for col in exclude_columns:
    if col in categorical.columns:
        drop_list.append(col)

# Drop the columns from the drop_list
categorical.drop(drop_list, axis=1, inplace=True)
```

## Identify columns that over 85% missing values

```python
# Calculate the percentage of missing values for each column
missing_percentages = (categorical.isna().sum() / len(categorical)) * 100

# Filter columns with over 85% missing values
over_85_percent = missing_percentages[missing_percentages > 85]

# Get the column names as a list
columns_to_drop = list(over_85_percent.index)

print("Columns with over 85% missing values:", columns_to_drop)
```

```python
# no columns with over 85% missing values
missing_percentages
```

```python
# Get value counts for each categorical column
for column in categorical:
    print("Value counts for column '{}':".format(column))
    print(data[column].value_counts())
    print()
```

```python
# a lot of " " --> Turn them into NaN
categorical_trans = categorical.replace(" ", np.nan) 

# Calculate the percentage of missing values for each column
missing_percentages_2 = (categorical_trans.isna().sum() / len(categorical_trans)) * 100

# Filter columns with over 85% missing values
over_85_percent_2 = missing_percentages_2[missing_percentages_2 > 85]

# Get the column names as a list
columns_to_drop_2 = list(over_85_percent_2.index)

print("Columns with over 85% missing values:", columns_to_drop_2)
```

## Remove those columns from the dataframe

```python
categorical_trans.drop(columns_to_drop_2, axis=1, inplace=True)
categorical_trans.shape
```

## Reduce the number of categories in the column GENDER. The column should only have either "M" for males, "F" for females, and "other" for all the rest

- Note that there are a few null values in the column. We will first replace those null values using the code below:

```python
categorical_trans['gender'].value_counts()
```

```python
# Replace null values with 'F'
categorical_trans['gender'] = categorical_trans['gender'].fillna('F')
categorical_trans['gender'].value_counts()
```

```python
# Reduce categories in the 'GENDER' column
categorical_trans['gender'] = categorical_trans['gender'].apply(lambda x: 'M' if x == 'M' else 'F' if x == 'F' else 'other')
categorical_trans['gender'].value_counts()
```
